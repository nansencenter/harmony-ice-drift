```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
lonlat_crs = ccrs.PlateCarree()
stere_crs = ccrs.Stereographic(central_latitude=90,
                              central_longitude=-45,
                              true_scale_latitude=60)
```

```python
n = np.load('field_20190103T180000Z.npz')
x1, y1 = n['x'] + n['u'], n['y'] + n['v']
lon0, lat0, _ = lonlat_crs.transform_points(stere_crs,
                                            n['x'],
                                            n['y']).T
lon1, lat1, _ = lonlat_crs.transform_points(stere_crs,
                                            x1,
                                            y1).T
ulon, vlat = lon1 - lon0, lat1 - lat0
```

```python
stp = 100
plt.figure()
plt.quiver(lon0[::stp], lat0[::stp],
           ulon[::stp], vlat[::stp],
           angles='xy',
           scale_units='xy',
           scale=0.000001)
plt.show()
```

```python
stp = 100
plt.figure()
plt.quiver(n['x'][::stp], n['y'][::stp],
           n['u'][::stp], n['v'][::stp],
           angles='xy',
           scale_units='xy',
           scale=0.000001)
plt.show()
```
